```python
#Librerias implementadas
import scipy.io
import copy
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import time
import pickle
import scipy.io as sio
from sklearn import svm
from sklearn import linear_model
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn import tree
from sklearn.cluster import KMeans
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix, precision_score, f1_score, recall_score
```

```python
#funciones para almacenar el modelo.

def guardarModelo(filename,modelo):
    pickle.dump(modelo, open(filename+".npy", 'wb'))
def cargarModelo(filename):
    loaded_model = pickle.load(open(filename+".npy", 'rb'))
    return loaded_model
```

```python
#datos_tesis es una varible que guarda el nombre del archivo csv para su implementacion.

def cargardatos(datos_tesis):

    datos=pd.read_csv(datos_tesis, sep=',')
    datos=datos.drop(['LOGICA COMPUTACIONAL','PROGRAMACION I','PROGRAMACION II','PROGRAMACION III'], axis=1)
    #debido al desvalance de la categoria aceptable se asume que no tendra un resultado optimo evidenciado en la matriz de confusion.
    #datosentrenamiento=pd.read_csv(datos, sep=',')
    evaluacion=[]
    cont1=10
    cont2=10
    cont3=1
    for i in datos.index:
        if datos["CLASE"][i]=="bueno" and cont1!=0:
            evaluacion.append(i)
            cont1=cont1-1
        if datos["CLASE"][i]=="sobresaliente" and cont2!=0:
            evaluacion.append(i) 
            cont2=cont2-1 
        if datos["CLASE"][i]=="aceptable" and cont3!=0:
            evaluacion.append(i) 
            cont2=cont3-1 

#Asigno una lista (evaluacion), se elimina las columnas que no implementare para la evaluacion(logica,progrmacion I,II,II).
#implemento dos variables cont1,cont2 copn el numero de etiquetas  "bueno","sobresaliente" de la clase.
#Implemento un contador con el cual recorrer la lista y asignarle una de las dos etiquetas anteriores, hasta llenar la lista. 
#el numero total de datos de entreamiento sera 10- 1/4 de los datos totales obtenidos.
    datosentrenamiento=datos.drop(evaluacion,axis=0)

    datosevaluacion=datos.iloc[evaluacion]
    
    
    #llamado de la matriz (entrenamiento-evaluacion)/ nombramiento espesifico de las variables x,y.
    Xtraning = datosentrenamiento[['LECTURA CRITICA','MATEMATICA','INGLES','SOCIALES Y CIUDADANIA','CIENCIAS NATURALES']]
    Ytraningnumerico = datosentrenamiento['PROMEDIO PROGRAMACIONES']
    Ytraningnominal = datosentrenamiento['CLASE']
    Xtraning = Xtraning.to_numpy()

    Xeval = datosevaluacion[['LECTURA CRITICA','MATEMATICA','INGLES','SOCIALES Y CIUDADANIA','CIENCIAS NATURALES']]
    Yevalnumerico = datosevaluacion['PROMEDIO PROGRAMACIONES']
    Yevalnominal = datosevaluacion['CLASE']
    Xeval = Xeval.to_numpy()
    return(Xtraning,Ytraningnumerico,Ytraningnominal,Xeval,Yevalnumerico,Yevalnominal)
```

```python
#xtraining es una matriz numerica/ Ytraining es un vector/ alfa  es un decimal/ nombre_modelo es un string(nombre del modelo)

def train_lr( Xtraining,Ytrainingnumerico,alfa,nombre_modelo):
    #selectmodel = linear_model.Lasso(alpha=alfa) 
    selectmodel= LinearRegression()
    selectmodel.fit(Xtraining, Ytrainingnumerico)
    guardarModelo(nombre_modelo,selectmodel)
#xeval es una matriz numeroca con los valores de evaluacion/ Yevalnumerico es un vector con la clase numerica/Yevalnominal es una lista /Ytrainingnominal es el vector de entrenamiento(una lista que guerda las clases de los valores de entrenamiento) 
#nombre del modelo/archivo_resultado es una variable de tipo texto  que guarda el nombre del archivo deonde se van a guardar los resultados.    
def evaluar_modelo(Xeval,Yevalnumerico, Yevalnominal,Ytrainingnominal,nombre_modelo, archivo_resultados):
    modeloRegresion=cargarModelo(nombre_modelo)
#Ejecucion de las predicciones
    pre=modeloRegresion.predict(Xeval)
# Medidas de desempe√±o del modelo(presicion, covertura, f-medidas,matriz de confusion.)

    lista_p=[]
    for i in pre: 

        if i<3: 
            lista_p.append("aceptable")

        if i >3 and i <=3.7:
             lista_p.append("bueno")

        if i >=3.7 and i <5.0:

            lista_p.append("sobresaliente")
            
#matriz de confucion,precicion
    x_true=Yevalnominal.tolist();
    x_pred=lista_p
    conf_mat=confusion_matrix(x_true,x_pred,labels=["aceptable","bueno","sobresaliente"])
    precision=precision_score(x_true, x_pred, average='macro',zero_division=1)
    recall=recall_score(x_true, x_pred, average='macro',zero_division=1)
    f_score=f1_score(x_true, x_pred, average='macro',zero_division=1) 
    labels=["aceptable","bueno","sobresaliente"]

    print(conf_mat)
    print(precision)
    print(recall)
    print(f_score)
    archivo1=open(archivo_resultados+".txt","w")
    archivo1.write(str(conf_mat)+"\n")
    archivo1.write("\n"+str(precision)+"\n")
    archivo1.write(str(recall)+"\n")
    archivo1.write(str(f_score)+"\n")
    archivo1.write(str(labels)+"\n")
    archivo1.close()
    
def main():
    Xtraning,Ytraningnumerico,Ytraningnominal,Xeval,Yevalnumerico,Yevalnominal=cargardatos('Datos-tesis-clasificador.csv')
    print(Xtraning)
    print(Ytraningnumerico)
    print(Ytraningnominal)
    print(Xeval)
    print(Yevalnumerico)
    print(Yevalnominal)
    alfa=0.1
    nombre_modelo="Regresion"
    archivo_resultados="Resultados_obtenidos"
    train_lr( Xtraning,Ytraningnumerico,alfa,nombre_modelo)
    evaluar_modelo(Xeval,Yevalnumerico, Yevalnominal,Ytraningnominal,nombre_modelo, archivo_resultados)

    
```

```python
main()
```

```python
def train_SVM( Xtraining,Ytrainingnumerico,alfa,nombre_modelo,Ytraningnominal):
    #selectmodel = linear_model.Lasso(alpha=alfa) 
    modelo_svm=svm.SVC(gamma='auto')
    modelo_svm.fit(Xtraining,Ytraningnominal)
    guardarModelo(nombre_modelo, modelo_svm)
#xeval es una matriz numeroca con los valores de evaluacion/ Yevalnumerico es un vector con la clase numerica/Yevalnominal es una lista /Ytrainingnominal es el vector de entrenamiento(una lista que guerda las clases de los valores de entrenamiento) 
#nombre del modelo/archivo_resultado es una variable de tipo texto  que guarda el nombre del archivo deonde se van a guardar los resultados.    
def evaluar_SVM(Xeval,Yevalnumerico, Yevalnominal,Ytrainingnominal,nombre_modelo, archivo_resultados):
    modelo_svm=cargarModelo(nombre_modelo)
#Ejecucion de las predicciones
    pre=modelo_svm.predict(Xeval)

   
            
#matriz de confucion,precicion
    x_true=Yevalnominal.tolist();
    x_pred=pre.tolist()
    print(x_pred)
    print(x_true)
    conf_mat=confusion_matrix(x_true,x_pred,labels=["aceptable","bueno","sobresaliente"])
    precision=precision_score(x_true, x_pred, average='macro',zero_division=1)
    recall=recall_score(x_true, x_pred, average='macro',zero_division=1)
    f_score=f1_score(x_true, x_pred, average='macro',zero_division=1) 
    labels=["aceptable","bueno","sobresaliente"]

    print(conf_mat)
    print(precision)
    print(recall)
    print(f_score)
    archivo1=open(archivo_resultados+".txt","w")
    archivo1.write(str(conf_mat)+"\n")
    archivo1.write("\n"+str(precision)+"\n")
    archivo1.write(str(recall)+"\n")
    archivo1.write(str(f_score)+"\n")
    archivo1.write(str(labels)+"\n")
    archivo1.close()
    
def main_svm():
    Xtraning,Ytraningnumerico,Ytraningnominal,Xeval,Yevalnumerico,Yevalnominal=cargardatos('Datos-tesis-clasificador.csv')
    print(Xtraning)
    print(Ytraningnumerico)
    print(Ytraningnominal)
    print(Xeval)
    print(Yevalnumerico)
    print(Yevalnominal)
    alfa=0.1
    nombre_modelo="Support Vector Machine"
    archivo_resultados="Resultados-obtenidos-Support Vector Machine"
    train_SVM( Xtraning,Ytraningnumerico,alfa,nombre_modelo,Ytraningnominal)
    evaluar_SVM(Xeval,Yevalnumerico, Yevalnominal,Ytraningnominal,nombre_modelo, archivo_resultados)

    
```
